```python
# Create plots with pre-defined labels.
fig, ax = plt.subplots()
ax.plot(df_random['n'], df_random['ins'], 'b--', label='Insertion Sort')
ax.plot(df_random['n'], df_random['bub'], 'r:', label='Bubble Sort')
#ax.plot(df_random['n'],df_random['ins'], 'k', label='Total message length')

legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')

# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('C0')

plt.show()
```
